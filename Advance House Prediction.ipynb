# define metro cities

```python
metro_cities = ['Ahmedabad', 'Bangalore', 'Chennai', 'Delhi', 'Hyderabad', 'Kolkata', 'Mumbai', 'Pune', 'Gurgaon', 'Chandigarh']
```

# Taking cities from the full address

```python
train_data['city'] = train_data['ADDRESS'].str.split(",").str[-1]
test_data['city'] = test_data['ADDRESS'].str.split(",").str[-1]
```

```python
train_data.head()
```

# Taking 1 in place of metro cities and 0 for non metro cities 

```python
def set_metro_city(row, metro_city):
    if metro_city == 1:
        if row['city'] in metro_cities:
            return 1
        else:
            return 0

train_data['metro_city'] = train_data.apply(set_metro_city, args=([1]), axis=1)
test_data['metro_city'] = test_data.apply(set_metro_city, args=([1]), axis=1)
```
